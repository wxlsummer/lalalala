```python
import os
import wave
import librosa
import scipy.io.wavfile as wav
import numpy as np
 
def add_noise(data):
    wn = np.random.normal(0,500,len(data))
    data_noise = np.where(data != 0.0, data + 0.8 * wn, 0.0).astype(np.int16)
    return data_noise


name="sample.wav"
path_noise="WN_all.wav"

print("add noise File Name is ", name)
fs,data = wav.read(name)
data_noise = add_noise(data)

wav.write(path_noise,fs, data_noise)


```
